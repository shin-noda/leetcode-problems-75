```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def searchBST(self, root, val):
        """
        :type root: Optional[TreeNode]
        :type val: int
        :rtype: Optional[TreeNode]
        """

        self.subtree_node = None

        def find(node):
            # Base case
            if not node or self.subtree_node:
                return

            if node.val == val:
                self.subtree_node = node
                return

            find(node.left)
            find(node.right)

        find(root)

        return self.subtree_node
```
