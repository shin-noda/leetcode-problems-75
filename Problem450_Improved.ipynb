```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def deleteNode(self, root, key):
        """
        :type root: Optional[TreeNode]
        :type key: int
        :rtype: Optional[TreeNode]
        """

        # Use a dummy to handle the deleting the root case
        dummy = TreeNode()
        dummy.left = root
        curr = root
        prev = dummy

        # Search for the node and its parent
        while curr and curr.val != key:
            prev = curr

            if key < curr.val:
                curr = curr.left

            else:
                curr = curr.right

        # Didn't find the node, so no deletion
        if not curr:
            return dummy.left

        # Determine what replaces curr
        replacement = None
        if not curr.right:
            # If there's no right side, the left side just moves up
            replacement = curr.left

        else:
            # Find leftmost of the right subtree
            leftmost = curr.right
            while leftmost.left:
                leftmost = leftmost.left

            # Graft the left subtree onto the leftmost node
            leftmost.left = curr.left
            replacement = curr.right

        # Connect parent to the replacement
        if prev.left == curr:
            prev.left = replacement

        else:
            prev.right = replacement

        return dummy.left
```
