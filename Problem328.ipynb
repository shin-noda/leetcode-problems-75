```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def oddEvenList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        # Check if head is empty or only 1 or 2 nodes.
        if not head or not head.next or not head.next.next:
            return head

        # A sentinel node
        dummy = ListNode()
        dummy.next = head

        odds = head
        evens = head.next

        # Since we checked the condition above, it's guaranteed to have head.next
        # Note: head.next.next could be None.
        first_even = head.next

        # Increment the head
        head = head.next.next

        i = 3
        while head:
            # It's an odd node
            if i % 2 == 1:
                odds.next = head
                odds = odds.next

            # It's an even node
            else:
                evens.next = head
                evens = evens.next

            head = head.next
            i += 1

        # Combine odd and even nodes
        odds.next = first_even

        # Cut the loop at the end of evens
        evens.next = None

        return dummy.next
```
