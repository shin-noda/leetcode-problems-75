```python
class Solution(object):
    def equalPairs(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """

        count = 0
        n = len(grid)

        # Count occurrences of each row
        # We use a tuplet because lists cannot be keys in a dictionary
        row_counts = {}
        for row in grid:
            row_tuple = tuple(row)
            row_counts[row_tuple] = row_counts.get(row_tuple, 0) + 1

        # Extract each column and check if it exists in our row map
        for j in range(n):
            col = []
            for i in range(n):
                col.append(grid[i][j])

            # Convert a list to a tuple
            col_tuple = tuple(col)

            if col_tuple in row_counts:
                # e.g. if a row appears 3 times and a matching column appears once,
                # that's 3 pairs.
                count += row_counts[col_tuple]

        return count
```

```python
solution = Solution()
```

```python
grid = [[3,2,1],[1,7,6],[2,7,7]]

print(solution.equalPairs(grid))
```
