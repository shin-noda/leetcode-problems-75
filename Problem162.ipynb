```python
class Solution(object):
    def findPeakElement(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        n = len(nums)
        left = 0
        right = len(nums) - 1

        while left <= right:
            mid = (left + right) // 2

            # Check the left element
            if mid + 1 < n and nums[mid] < nums[mid + 1]:
                # We climb the hill to the left
                left = mid + 1

            elif mid > 0 and nums[mid] < nums[mid - 1]:
                # We climb the hill to the right
                right = mid - 1

            else:
                return mid
```

```python
solution = Solution()
```

```python
nums = [1,2,3,1]

print(solution.findPeakElement(nums))
```
