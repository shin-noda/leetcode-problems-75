```python
class Solution(object):
    def uniquePaths(self, m, n):
        """
        :type m: int
        :type n: int
        :rtype: int
        """

        dp = [0] * n

        # Update the first row
        for i in range(n):
            dp[i] = 1

        # Fill the rest of the rows
        for _ in range(1, m):
            # We start from the 2nd column because the leftmost cells are always 1.
            for i in range(1, n):
                # We add the left and the top cells.
                # Left: dp[i - 1]
                # Top: dp[i]
                dp[i] += dp[i - 1]

        return dp[-1]
```

```python
solution = Solution()
```

```python
m = 3
n = 7

print(solution.uniquePaths(m, n))
```
