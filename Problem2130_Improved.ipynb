```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def pairSum(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: int
        """

        # This solution has space complexity of O(1).
        # Idea:
        # 1. Use the fast and slow pointer to find the middle point
        # 2. While traversing nodes, we reverse nodes until the middle
        # 3. We traverse the reversed nodes and the remaining half nodes
        # at the same time to calculate the max sum.

        # Check if nodes exist
        if not head or not head.next:
            return head

        slow = head
        fast = head.next.next

        prev = None
        curr = head
        while fast and fast.next:
            # The fast and slow pointer method
            slow = slow.next
            fast = fast.next.next

            # Have a backup
            next_node = curr.next

            # Connect the current node to the previous one
            curr.next = prev

            # Update the previous node
            prev = curr

            # Proceed to the next node
            curr = next_node

        # We need to forward slow once
        slow = slow.next

        max_val = slow.val + curr.val

        # Forward the slow
        if slow:
            slow = slow.next

        # We have prev as the staring point of the reverse node
        while slow and prev:
            max_val = max(max_val, slow.val + prev.val)

            slow = slow.next
            prev = prev.next

        return max_val
```
