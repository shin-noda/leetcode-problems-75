```python
class Solution(object):
    def mergeAlternately(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: str
        """

        # Check if both strings are empty
        if not word1 and not word2:
            return ""

        merged_string = ""
        n1 = len(word1)
        n2 = len(word2)
        index = 0

        while index < n1 and index < n2:
            merged_string = merged_string + word1[index] + word2[index]
            index += 1

        if n1 < n2:
            merged_string = merged_string + word2[index:]

        else:
            merged_string = merged_string + word1[index:]

        return merged_string
```

```python
solution = Solution()
```

```python
word1 = "abc"
word2 = "pqr"

print(solution.mergeAlternately(word1, word2))
```
