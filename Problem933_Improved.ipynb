```python
class RecentCounter(object):

    def __init__(self):
        self.queue = []


    def ping(self, t):
        """
        :type t: int
        :rtype: int
        """

        # Add the new timestamp to our list
        self.queue.append(t)

        # The oldest elements are at index 0
        while self.queue[0] < t - 3000:
            self.queue.pop(0)

        # The length of the list is now exactly the count of valid pings
        return len(self.queue)
```
