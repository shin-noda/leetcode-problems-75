```python
class Solution(object):
    def maxVowels(self, s, k):
        """
        :type s: str
        :type k: int
        :rtype: int
        """

        # Check if k is larger than the length of s
        if k > len(s):
            return 0

        left = 0
        curr = 0
        max_vowels = 0

        # Use a set for O(1) look ups
        vowels = set("aeiou")

        for i in range(k):
            if s[i] in vowels:
                max_vowels += 1
                curr = max_vowels

        for i in range(k, len(s)):
            if s[left] in vowels:
                curr -= 1

            if s[i] in vowels:
                curr += 1

            max_vowels = max(max_vowels, curr)
            left += 1

        return max_vowels
```

```python
solution = Solution()
```

```python
s = "abciiidef"
k = 3

print(solution.maxVowels(s, k))
```
