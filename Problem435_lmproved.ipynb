```python
from collections import deque

class Solution(object):
    def eraseOverlapIntervals(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: int
        """

        if not intervals:
            return 0

        # Sort by end time (x[1])
        intervals.sort(key=lambda x: x[1])

        count = 0

        # Track the end of the last interval we added to our "non-overlapping" set
        prev_end = intervals[0][1]

        # Start from the second interval
        for i in range(1, len(intervals)):
            curr_start = intervals[i][0]
            curr_end = intervals[i][1]

            if curr_start < prev_end:
                # Overlap. We must remove this one
                # Since we sorted by end time, the current one ends later
                # or at the same time as prev_end, so we always discard curr
                count += 1

            else:
                # No overlap. Update our tracker to the new end time
                prev_end = curr_end

        return count
```

```python
solution = Solution()
```

```python
intervals = [[1,2],[2,3],[3,4],[1,3]]

print(solution.eraseOverlapIntervals(intervals))
```
