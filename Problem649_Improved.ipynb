```python
class Solution(object):
    def predictPartyVictory(self, senate):
        """
        :type senate: str
        :rtype: str
        """

        queue = list(senate)
        winner = None
        pointer = 0

        while pointer < len(queue) and not winner:
            senator = queue[pointer]

            if senator == "R":
                is_victory = True
                if not queue:
                    winner = "Radiant"

                else:
                    for i in range(pointer, len(queue)):
                        # Ban Dire's rights
                        if queue[i] == "D":
                            queue[i] = ""
                            is_victory = False
                            break

                    if is_victory:
                        winner = "Radiant"

            elif senator == "D":
                is_victory = True
                if not queue:
                    winner = "Dire"

                else:
                    for i in range(pointer, len(queue)):
                        # Ban Radiant's rights
                        if queue[i] == "R":
                            queue[i] = ""
                            is_victory = False
                            break

                    if is_victory:
                        winner = "Dire"

            # Bring the senator back to the queue
            queue.append(senator)

            pointer += 1

        return winner
```

```python
solution = Solution()
```

```python
senate = "RD"

print(solution.predictPartyVictory(senate))
```
