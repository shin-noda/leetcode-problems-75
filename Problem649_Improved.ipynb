```python
class Solution(object):
    def predictPartyVictory(self, senate):
        """
        :type senate: str
        :rtype: str
        """

        queue = list(senate)
        pointer = 0
        r_bans = 0
        d_bans = 0
        rs = 0
        ds = 0

        # Count the number of Radiants and Dires
        for senator in queue:
            if senator == "R":
                rs += 1

            else:
                ds += 1

        while pointer < len(queue):
            senator = queue[pointer]

            if senator == "R":
                if r_bans > 0:
                    r_bans -= 1

                else:
                    d_bans += 1
                    ds -= 1

                    if ds <= 0:
                        return "Radiant"

                    else:
                        queue.append(senator)

            else:
                if d_bans > 0:
                    d_bans -= 1

                else:
                    r_bans += 1
                    rs -= 1

                    if rs <= 0:
                        return "Dire"

                    else:
                        queue.append(senator)

            pointer += 1
```

```python
solution = Solution()
```

```python
senate = "RD"

print(solution.predictPartyVictory(senate))
```
