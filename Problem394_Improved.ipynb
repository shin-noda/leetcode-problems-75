```python
class Solution(object):
    def decodeString(self, s):
        """
        :type s: str
        :rtype: str
        """

        stack = []
        cur_num = 0
        cur_str = ''

        for char in s:
            if char == "[":
                # Save the state before the bracket
                stack.append(cur_str)
                stack.append(cur_num)

                # Reset for the content inside
                cur_str = ''
                cur_num = 0

            elif char == ']':
                # Pop the multiplier and the previous string
                num = stack.pop()
                prev_str = stack.pop()
                cur_str = prev_str + (num * cur_str)

            elif char.isdigit():
                # Build the number (handles multi-digits like '100')
                cur_num = cur_num * 10 + int(char)

            else:
                cur_str += char

        return cur_str
```

```python
solution = Solution()
```

```python
s = "3[a]2[bc]"

print(solution.decodeString(s))
```

```python
s = "3[a2[c]]"

print(solution.decodeString(s))
```
