```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def leafSimilar(self, root1, root2):
        """
        :type root1: Optional[TreeNode]
        :type root2: Optional[TreeNode]
        :rtype: bool
        """

        # The generator stays the same
        def getLeaves(node):
            if node:
                if not node.left and not node.right:
                    yield node.val

                # Recursive Case: manually pass up values from children
                if node.left:
                    for leaf in getLeaves(node.left):
                        yield leaf

                if node.right:
                    for leaf in getLeaves(node.right):
                        yield leaf


        # We create two "iterator" objects
        it1 = getLeaves(root1)
        it2 = getLeaves(root2)

        while True:
            # Manually pull the next leaf from each tree
            # We use a unique sentinel (None) to signal the end
            v1 = next(it1, None)
            v2 = next(it2, None)

            # If the values don't match, we stop immediately.
            if v1 != v2:
                return False

            # If both reached the end (None) at the same time, they are identical
            # v2 is also None because v1 == v2
            if v1 is None:
                return True
```
