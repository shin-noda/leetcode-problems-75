```python
class Solution(object):
    def minReorder(self, n, connections):
        """
        :type n: int
        :type connections: List[List[int]]
        :rtype: int
        """

        # Idea: Start from city 0 and walk outward.
        # Count how many edges are pointing the wrong way

        # Build adjacency list manually
        # Each entry will be [neighbor, cost]
        adj = [[] for _ in range(n)]
        for u, v in connections:
            # Original: away from 0
            adj[u].append((v, 1))

            # Artificial: toward 0
            adj[v].append((u, 0))

        count = 0

        # BFS using a list and a pointer
        queue = [0]
        visited = [False] * n
        visited[0] = True

        head = 0
        while head < len(queue):
            curr = queue[head]
            head += 1

            for neighbor, cost in adj[curr]:
                # This if for preventing revisiting nodes and going in circles.
                if not visited[neighbor]:
                    visited[neighbor] = True

                    # If cost is 1, the arrow was pointing away from 0
                    count += cost
                    queue.append(neighbor)

        return count
```

```python
solution = Solution()
```

```python
n = 6
connections = [[0,1],[1,3],[2,3],[4,0],[4,5]]

print(solution.minReorder(n, connections))
```
