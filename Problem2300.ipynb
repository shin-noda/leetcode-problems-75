```python
class Solution(object):
    def successfulPairs(self, spells, potions, success):
        """
        :type spells: List[int]
        :type potions: List[int]
        :type success: int
        :rtype: List[int]
        """

        def binarySearch(val):
            left = 0
            right = len(potions) - 1

            while left <= right:
                mid = (left + right) // 2

                if potions[mid] * val >= success:
                    right = mid - 1

                else:
                    left = mid + 1

            return right + 1

        # We sort positions: O(N log N)
        potions.sort()

        results = []
        for i in range(len(spells)):
            # Avoid tmp =[x * sp[i][1] for x in potions] because it's O(N)
            result = binarySearch(spells[i])
            results.append(len(potions) - result)

        return results
```

```python
solution = Solution()
```

```python
spells = [5,1,3]
potions = [1,2,3,4,5]
success = 7

print(solution.successfulPairs(spells, potions, success))
```
