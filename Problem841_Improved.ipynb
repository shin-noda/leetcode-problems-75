```python
class Solution(object):
    def canVisitAllRooms(self, rooms):
        """
        :type rooms: List[List[int]]
        :rtype: bool
        """

        # Rooms we have actually entered
        visited = {0}

        # Keys we have found but haven't used to enter a room yet
        stack = [0]

        while stack:
            current_room = stack.pop()

            for key in rooms[current_room]:
                if key not in visited:
                    visited.add(key)
                    stack.append(key)

        # If the number of visited rooms matches the total rooms, we made it
        return len(visited) == len(rooms)
```

```python
solution = Solution()
```

```python
rooms = [[1],[2],[3],[]]

print(solution.canVisitAllRooms(rooms))
```

```python
rooms = [[1,3],[3,0,1],[2],[0]]

print(solution.canVisitAllRooms(rooms))
```
