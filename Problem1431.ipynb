```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        """
        :type candies: List[int]
        :type extraCandies: int
        :rtype: List[bool]
        """

        results = []
        max_c  = max(candies)

        for c in candies:
            results.append(c + extraCandies >= max_c)

        return results
```

```python
solution = Solution()
```

```python
candies = [2,3,5,1,3]
extraCandies = 3

print(solution.kidsWithCandies(candies, extraCandies))
```

```python
candies = [4,2,1,1,2]
extraCandies = 1

print(solution.kidsWithCandies(candies, extraCandies))
```
