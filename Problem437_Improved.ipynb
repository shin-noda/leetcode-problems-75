```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def pathSum(self, root, targetSum):
        """
        :type root: Optional[TreeNode]
        :type targetSum: int
        :rtype: int
        """

        # Use Backtracking to achieve O(N) complexity

        # self.m stores {prefix sum: count}
        # We initialize it with {0: 1} to account for paths that
        # start exactly from the root.
        self.m = {0: 1}
        self.count = 0

        def dive(node, current_sum):
            # Base case
            if not node:
                return

            # Update the running prefix sum
            current_sum += node.val

            # Check the complement:
            # If (current_sum - targetSum) is in the map, it means
            # a path exists somewhere above that makes the middle path valid
            self.count += self.m.get(current_sum - targetSum, 0)

            # Add current_sum to map so children can use it as a starting point
            self.m[current_sum] = self.m.get(current_sum, 0) + 1

            # Explore downward
            dive(node.left, current_sum)
            dive(node.right, current_sum)

            # Backtrack: Remove current_sum so it doesnt' lean into other branches
            self.m[current_sum] -= 1

        dive(root, 0)
        return self.count
```
