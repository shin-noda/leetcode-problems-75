```python
class RecentCounter(object):

    def __init__(self):
        self.queue = []


    def ping(self, t):
        """
        :type t: int
        :rtype: int
        """

        self.queue.append(t)

        last = t - 3000

        count = 0
        i = len(self.queue)
        while i > 0:
            if self.queue[-(count + 1)] >= last:
                count += 1

            else:
                break

            i -= 1

        return count
```
