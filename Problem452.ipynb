```python
from collections import deque

class Solution(object):
    def findMinArrowShots(self, points):
        """
        :type points: List[List[int]]
        :rtype: int
        """

        count = 0

        # Sort points by x[1] (x_end) and
        # shoot an arrow aiming at each x_end
        points.sort(key=lambda x: x[1])

        dq = deque(points)
        prev_start, prev_end = dq.popleft()
        count += 1

        while dq:
            curr_start, curr_end = dq.popleft()

            # It hits the arrow
            if curr_start <= prev_end <= curr_end:
                continue

            else:
                # Update the amining point
                prev_end = curr_end
                count += 1

        return count
```

```python
solution = Solution()
```

```python
points = [[10,16],[2,8],[1,6],[7,12]]

print(solution.findMinArrowShots(points))
```
