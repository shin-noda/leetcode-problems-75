```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def rightSideView(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: List[int]
        """

        # Check if there are nodes
        if not root:
            return []

        queue = [root]
        vals = []

        pointer = 0
        while pointer < len(queue):
            # Append the rightmost value
            vals.append(queue[-1].val)

            n = len(queue) - pointer
            for i in range(pointer, len(queue)):
                node = queue[i]

                # Since we want to get node values looking from the right side,
                # we append the left node first.
                if node.left:
                    queue.append(node.left)

                if node.right:
                    queue.append(node.right)

                # Forward the pointer
                pointer += 1

        return vals
```
