```python
class Solution(object):
    def pivotIndex(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # Pivot Index: the index where the sum of all the numbers
        # strictly to the left of the index is equal to the sum of all the numbers
        # strictly to the index's right.

        sum_val = sum(nums)
        left_sum = 0

        for i in range(len(nums)):
            right_sum = sum_val - left_sum - nums[i]

            if left_sum == right_sum:
                return i

            # Increment the left sum
            left_sum += nums[i]

        # There is no pivot index
        return -1
```

```python
solution = Solution()
```

```python
nums = [1,7,3,6,5,6]

print(solution.pivotIndex(nums))
```
