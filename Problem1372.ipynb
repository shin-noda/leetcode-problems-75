```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def longestZigZag(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """


        # Note: move = 0 -> Left, move = 1 -> Right
        def moveZigZag(node, move, count):
            # Base Case
            if not node:
                return count

            # Update the count
            count = count + 1

            if move == 0:
                return max(
                    moveZigZag(node.left, 0, 0),
                    moveZigZag(node.right, 1, count)
                )

            else:
                return max(
                    moveZigZag(node.left, 0, count),
                    moveZigZag(node.right, 1, 0)
                )

        return max(moveZigZag(root.left, 0, 0), moveZigZag(root.right, 1, 0))
```
