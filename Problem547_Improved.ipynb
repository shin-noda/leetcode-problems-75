```python
class Solution(object):
    def findCircleNum(self, isConnected):
        """
        :type isConnected: List[List[int]]
        :rtype: int
        """

        # Use Union-Find (Disjoint Set)

        n = len(isConnected)

        # e.g. If n = 3, then parent = [0, 1, 2]
        parent = list(range(n))

        # Start with n provinces
        self.count = n

        def find(i):
            if parent[i] == i:
                return i

            # Path compression
            parent[i] = find(parent[i])

            return parent[i]

        def union(i, j):
            root_i = find(i)
            root_j = find(j)

            if root_i != root_j:
                parent[root_i] = root_j

                # Two provinces merged into one
                self.count -= 1

        for i in range(n):
            # Only need to check the upper triangle
            # Note: the matrix is matrix, so we start from i + 1,
            # ignore the diagonal line and below that.
            for j in range(i + 1, n):
                if isConnected[i][j] == 1:
                    union(i, j)

        return self.count
```

```python
solution = Solution()
```
