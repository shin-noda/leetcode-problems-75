```python
class Solution(object):
    def maxOperations(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """

        # Sort the array first
        nums.sort()

        # Use a two-pointer method
        left = 0
        right = len(nums) - 1
        operations = 0

        # Use two pointers to meet in the middle
        while left < right:
            current_sum = nums[left] + nums[right]

            if current_sum == k:
                # We found a pair
                operations += 1
                left += 1
                right -= 1

            elif current_sum < k:
                # Sum is too small, so move left pointer to a larger value
                left += 1

            else:
                # Sum is too big, so move right pointer to a smaller value
                right -= 1

        return operations
```

```python
solution = Solution()
```

```python
nums = [1,2,3,4]
k = 5

print(solution.maxOperations(nums, k))
```
