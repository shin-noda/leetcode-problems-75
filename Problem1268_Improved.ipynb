```python
class TrieNode(object):
    def __init__(self):
        self.children = {}

        # Pre-store the top 3 suggestions at this node
        self.suggestions = []


class Solution(object):
    def suggestedProducts(self, products, searchWord):
        """
        :type products: List[str]
        :type searchWord: str
        :rtype: List[List[str]]
        """

        # Sort products so we process then in alphabetical order
        products.sort()
        root = TrieNode()

        # Build the Trie
        for product in products:
            node = root

            for char in product:
                if char not in node.children:
                    node.children[char] = TrieNode()

                node = node.children[char]

                # If we haven't found 3 suggestions for this prefix yet, add it
                if len(node.suggestions) < 3:
                    node.suggestions.append(product)

        # Search is now incredibly fast
        result = []
        node = root

        for char in searchWord:
            # If the character exists, move down; otherwise, node becomes None
            if node:
                node = node.children.get(char)

            # If node exists, grab the pre-calculated suggestions
            if node:
                result.append(node.suggestions)

            else:
                result.append([])

        return result
```

```python
solution = Solution()
```

```python
products = ["mobile","mouse","moneypot","monitor","mousepad"]
searchWord = "mouse"

print(solution.suggestedProducts(products, searchWord))
```
