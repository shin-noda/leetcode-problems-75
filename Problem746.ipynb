```python
class Solution(object):
    def minCostClimbingStairs(self, cost):
        """
        :type cost: List[int]
        :rtype: int
        """

        # We added 1 to represent the goal
        dp = [0] * (len(cost) + 1)

        # Update the first and the second elements of dp
        dp[0] = cost[0]
        dp[1] = cost[1]

        for i in range(2, len(cost) + 1):
            if i != len(cost):
                dp[i] = cost[i] + min(dp[i - 1], dp[i - 2])

            # This is for the top floor
            else:
                dp[i] = min(dp[i - 1], dp[i - 2])


        return dp[-1]
```

```python
solution = Solution()
```

```python
cost = [10,15,20]

print(solution.minCostClimbingStairs(cost))
```

```python
cost = [1,100,1,1,1,100,1,1,100,1]

print(solution.minCostClimbingStairs(cost))
```
