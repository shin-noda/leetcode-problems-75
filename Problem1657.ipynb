```python
class Solution(object):
    def closeStrings(self, word1, word2):
        """
        :type word1: str
        :type word2: str
        :rtype: bool
        """

        # Time Complexity: O(N) Note: Tim Sort is O(NlogN) but there's only 26 chars max
        # so, it's constant.
        # Space Complexity: O(1) Note: Since max length is 26, it's constant.

        # Check if the lengths of word1 and word2 are the same
        if len(word1) != len(word2):
            return False

        # We count the number of occurrences of both string and
        # sort them in non-decreasing order.
        # If they match, it's true.
        m1 = {}
        m2 = {}

        for w1, w2 in zip(word1, word2):
            m1[w1] = m1.get(w1, 0) + 1
            m2[w2] = m2.get(w2, 0) + 1

        k1 = set(m1.keys())
        k2 = set(m2.keys())

        # Both keys should have the same value
        if k1 != k2:
            return False

        v1 = list(m1.values())
        v2 = list(m2.values())

        v1.sort()
        v2.sort()

        return v1 == v2
```

```python
solution = Solution()
```
