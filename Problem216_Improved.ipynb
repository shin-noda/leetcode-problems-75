```python
class Solution(object):
    def combinationSum3(self, k, n):
        """
        :type k: int
        :type n: int
        :rtype: List[List[int]]
        """

        self.count = 0
        self.tmp = []
        self.combinations = []

        def backtrack(index):
            # Pruning
            if self.count > n:
                return

            # Base case: success
            if len(self.tmp) == k and self.count == n:
                self.combinations.append(list(self.tmp))
                return

            for i in range(index + 1, 10):
                # Choose
                self.tmp.append(i)
                self.count += i

                # Explore
                backtrack(i)

                # Backtrack
                self.tmp.pop()
                self.count -= i

        backtrack(0)

        return self.combinations
```

```python
solution = Solution()
```

```python
k = 3
n = 7

print(solution.combinationSum3(k, n))
```
