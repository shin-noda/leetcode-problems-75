```python
class Solution(object):
    def dailyTemperatures(self, temperatures):
        """
        :type temperatures: List[int]
        :rtype: List[int]
        """

        answer = [0] * len(temperatures)

        stack = []
        last = 0

        for i in range(len(temperatures) -1, -1, -1):
            temp = temperatures[i]

            lost = 0
            counter = 1
            while stack and stack[-1][0] <= temp:
                item, prev_lost = stack.pop()
                counter += 1 + prev_lost
                lost += prev_lost

                if stack:
                    lost += 1

            if len(stack) != 0:
                if last >= len(stack):
                    answer[i] = counter

                else:
                    answer[i] = 1

            stack.append((temp, lost))
            last = len(stack)

        return answer
```

```python
solution = Solution()
```
