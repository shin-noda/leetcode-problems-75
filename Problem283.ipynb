```python
class Solution(object):
    def moveZeroes(self, nums):
        """
        :type nums: List[int]
        :rtype: None Do not return anything, modify nums in-place instead.
        """

        i = 0
        zero_counts = 0
        while i < len(nums):
            num = nums[i]

            if num == 0:
                nums.pop(i)
                zero_counts += 1

            else:
                i += 1

        # Add 0's at the end of the list
        for _ in range(zero_counts):
            nums.append(0)
```

```python
solution = Solution()
```

```python
nums = [0,1,0,3,12]

print(solution.moveZeroes(nums))
```

```python
nums = [0]

print(solution.moveZeroes(nums))
```
