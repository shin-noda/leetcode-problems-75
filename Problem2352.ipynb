```python
class Solution(object):
    def equalPairs(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """

        # Time Complexity: O(N^2)
        # Space Complexity: O(N)

        # We store the numbers in each row as a single number.
        # e.g. [3,1,2,2] -> 3122
        rows = []
        cols = []
        count = 0

        for i in range(len(grid)):
            r = []
            c = []
            for j in range(len(grid)):
                r.append(grid[i][j])
                c.append(grid[j][i])

            rows.append(r)
            cols.append(c)

        for r in rows:
            for c in cols:
                if r == c:
                    count += 1

        return count
```

```python
solution = Solution()
```

```python
grid = [[3,2,1],[1,7,6],[2,7,7]]

print(solution.equalPairs(grid))
```
