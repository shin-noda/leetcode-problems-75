```python
class MinHeap:
    def __init__(self):
        self.heap = []


    def insert(self, value):
        # Add to the end
        self.heap.append(value)
        self._heapify_up(len(self.heap) - 1)


    def extract_min(self):
        if not self.heap:
            return None

        if len(self.heap) == 1:
            return self.heap.pop()

        # Swap root with last
        min_val = self.heap[0]
        self.heap[0] = self.heap.pop()

        # Restore heap property
        self._heapify_down(0)

        return min_val


    def _heapify_up(self, index):
        while index > 0:
            parent = (index - 1) // 2

            if self.heap[index] <self.heap[parent]:
                self.heap[index], self.heap[parent] = self.heap[parent], self.heap[index]
                index = parent

            else:
                break


    def _heapify_down(self, index):
        size = len(self.heap)

        while True:
            left = 2 * index + 1
            right = 2 * index + 2
            smallest = index

            if left < size and self.heap[left] < self.heap[smallest]:
                smallest = left

            if right < size and self.heap[right] < self.heap[smallest]:
                smallest = right

            if smallest == index:
                break

            self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
            index = smallest


    def peek(self):
        if self.heap:
            return self.heap[0]

        return None


class Solution(object):
    def maxScore(self, nums1, nums2, k):
        """
        :type nums1: List[int]
        :type nums2: List[int]
        :type k: int
        :rtype: int
        """

        # Idea: The Fixed Minimum
        # If we decide that a specific value nums2[i] must be the minimum in our subsequence,
        # then we are only allowed to pick other indices where nums2[j] >= nums2[i].
        # To maximize the score for that specific minimum, we would naturally want to pick
        # the k largest corresponding values from nums1.

        # Time Complexity: O(N log N) for soring and O(N log K) for the heap operations
        # Space Complexity: O(N) to store the pairs and O(k) for the heap

        # Steps
        # 1. Pair nums1 and nums2
        # 2. Sort by nums2 in descending order
        # 3. Use a Min Heap for nums1
        # 4. Calculate the current sum
        # 5. Calculate the score

        # Step 1. Pair nums1 and nums2
        z = zip(nums1, nums2)

        # Step 2: Sort by nums2 in descending order
        pairs = sorted(z, key=lambda x: x[1], reverse=True)

        print(pairs)

        # Step 3: Use a Min Heap for nums1
        min_heap = MinHeap()
        n = len(pairs)
        current_sum = 0
        score = 0
        for i in range(n):
            # The Min Heap has less than k values
            if len(min_heap.heap) < k:
                # Step 4. Calculate the current sum
                current_sum += pairs[i][0]
                min_heap.insert(pairs[i][0])

            # The Min Heap has k values
            else:
                extracted = min_heap.extract_min()
                min_heap.insert(pairs[i][0])

                # Update the current_sum
                current_sum += pairs[i][0] - extracted

            # Step 5. Calculate the score
            if len(min_heap.heap) == k:
                score = max(score, current_sum * pairs[i][1])

        return score
```

```python
solution = Solution()
```

```python
nums1 = [1,3,3,2]
nums2 = [2,1,3,4]
k = 3

print(solution.maxScore(nums1, nums2, k))
```
