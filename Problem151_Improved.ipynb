```python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """

        # Splits any whitespace
        # Note: .split() and split(" ") behaves differently.
        words = s.split()

        return " ".join(words[::-1])
```

```python
solution = Solution()
```

```python
s = "the sky is blue"

print(solution.reverseWords(s))
```

```python
s = "  hello world  "

print(solution.reverseWords(s))
```
