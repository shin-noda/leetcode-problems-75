```python
class Solution(object):
    def gcdOfStrings(self, str1, str2):
        """
        :type str1: str
        :type str2: str
        :rtype: str
        """

        # Check if they share the same string
        if str1 + str2 != str2 + str1:
            return ""

        # GCD function
        def get_gcd(a, b):
            while b:
                a, b = b, a % b

            return a

        # Find the length and alice
        gcd_length = get_gcd(len(str1), len(str2))

        return str1[:gcd_length]
```

```python
solution = Solution()
```
