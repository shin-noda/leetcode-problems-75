```python
class Solution(object):
    def numTilings(self, n):
        """
        :type n: int
        :rtype: int
        """

        # We define domino and tromino tiles in the following way:
        # Vertical d tile like "|": [3]
        # horizontal d tile placed at the top "-": [1, 1]
        # horizontal d tile placed at the bottom "_": [2, 2]
        # T tile looks like "L": [3, 2]
        # T tile looks like the Greek letter Gamma "Γ": [3, 1]
        # T tile looks like a reverse L horizontally: [2, 3]
        # T tile looks like a reverse Γ horizontally: [1, 3]

        # We want in the each column to have the value of 3.

        # When we use the "L", we have to use either the "L" or the "Γ" to close it.

        # There are four ways to end the right most side:
        # flat like [3] or [1, 1] "|" and [2, 2] etc.,
        # dent at the bottom like "Γ" [3, 1],
        # dent at the top like "L" [3, 2], and
        # nothing (so that we use "-" and "_" next)

        # Valid combinations: (Note: we have to fill the "dent" first and
        # fills from the top for a tie-breaker after "|")
        # After flat like "|": "|", "L", "Γ", "-". (No "_")
        # After something like "L": the reverse "Γ", "-"
        # After something like "Γ": "_" and the reverse "L"

        tiles = {
            1: [[1, 1], [1, 3]],
            2: [[2, 2], [2, 3]],
            3: [[3, 0], [3, 1], [3, 2]]
        }

        # Check if n is 1
        if n == 1:
            return 1

        # Check if ne is 2
        if n == 2:
            return 2

        # Note: g_end is for "Γ" like ends with a dent at the bottom like a gamma.
        curr = 2
        prev1 = 2
        prev2 = 1

        l_end = 1
        g_end = 1

        for i in range(3, n + 1):
            curr += prev2 + l_end + g_end

            l_end += prev2
            g_end += prev2

            prev2 = prev1
            prev1 = curr

        # We are asked to return the answer it modulo 10^9 + 7.
        return curr % (pow(10, 9) + 7)
```

```python
solution = Solution()
```
