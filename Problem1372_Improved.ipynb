```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def longestZigZag(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """

        self.max_length = 0

        def dfs(node):
            if not node:
                # Return [-1, -1] because a null node doesn't contribute
                # to the edge count.
                return [-1, -1]

            left_child = dfs(node.left)
            right_child = dfs(node.right)

            # Current node's potential
            # To go left from here, we must have come from the child's right path
            can_go_left = left_child[1] + 1

            # To go right from here, we must have come from the child's left path
            can_go_right = right_child[0] + 1

            # Update the global record
            self.max_length = max(self.max_length, can_go_left, can_go_right)

            return [can_go_left, can_go_right]

        dfs(root)

        return self.max_length
```
