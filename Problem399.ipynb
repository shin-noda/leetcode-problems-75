```python
class Node(object):
    def __init__(self, val):
        self.val = val
        self.nodes = set()


class Solution(object):
    def calcEquation(self, equations, values, queries):
        """
        :type equations: List[List[str]]
        :type values: List[float]
        :type queries: List[List[str]]
        :rtype: List[float]
        """

        # Create an adjacency list
        self.adj = {}
        node_map = {}
        results = [-1.0] * len(queries)

        for i in range(len(equations)):
            op1, op2 = equations[i]
            val = values[i]

            if op1 not in self.adj:
                self.adj[op1] = {}

            self.adj[op1][op2] = (val, 1.0)

            if op2 not in self.adj:
                self.adj[op2] = {}

            self.adj[op2][op1] = (1.0, val)

            if op1 not in node_map:
                node1 = Node(op1)
                node_map[op1] = node1

            if op2 not in node_map:
                node2 = Node(op2)
                node_map[op2] = node2

            node_map[op1].nodes.add(node_map[op2])
            node_map[op2].nodes.add(node_map[op1])

        n = len(queries)

        # Solve all equation
        for i in range(n):
            op1, op2 = queries[i]

            if op1 not in node_map or op2 not in node_map:
                # The queries is undetermined
                continue

            self.result = 0.0

            if op1 == op2:
                self.result = 1.0

            else:
                self.traverse(node_map[op1], op2, set(), 1.0, 1.0)

            if self.result != 0.0:
                results[i] = self.result

        return results


    def traverse(self, root, dest, visited, numerator, denominator):
        # Base case
        if not root or root.val in visited:
            return

        visited.add(root.val)

        if root.val == dest:
            self.result = float(numerator) / denominator
            return

        nodes = root.nodes

        for node in nodes:
            num, den = self.adj[root.val][node.val]
            self.traverse(node, dest, visited, numerator * num, denominator * den)

        return
```

```python
solution = Solution()
```

```python
equations = [["a","b"],["b","c"]]
values = [2.0,3.0]
queries = [["a","c"],["b","a"],["a","e"],["a","a"],["x","x"]]

print(solution.calcEquation(equations, values, queries))
```
