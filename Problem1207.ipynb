```python
class Solution(object):
    def uniqueOccurrences(self, arr):
        """
        :type arr: List[int]
        :rtype: bool
        """

        # Create a hash map to count the frequency of numbers
        m = {}
        freq = set()

        for num in arr:
            m[num] = m.get(num, 0) + 1

        for key in m:
            val = m[key]
            if val in freq:
                return False

            freq.add(val)

        return True
```

```python
solution = Solution()
```

```python
arr = [1,2,2,1,1,3]

print(solution.uniqueOccurrences(arr))
```

```python
arr = [1,2]

print(solution.uniqueOccurrences(arr))
```
