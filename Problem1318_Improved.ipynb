```python
class Solution(object):
    def minFlips(self, a, b, c):
        """
        :type a: int
        :type b: int
        :type c: int
        :rtype: int
        """

        # We want to flip bits of a and b so that
        # a | b == c

        flips = 0

        while a or b or c:
            abit = a & 1
            bbit = b & 1
            cbit = c & 1

            # We need to flip both a and b when
            # a == 1 and b == 1
            if cbit == 0:
                if abit:
                    flips += 1

                if bbit:
                    flips += 1

            else:
                # if cbit == 1, we need at least one 1
                if abit == 0 and bbit == 0:
                    flips += 1

            a >>= 1
            b >>= 1
            c >>= 1

        return flips
```

```python
solution = Solution()
```

```python
a = 2
b = 6
c = 5

print(solution.minFlips(a, b, c))
```
