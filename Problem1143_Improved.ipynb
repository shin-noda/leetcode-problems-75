```python
class Solution(object):
    def longestCommonSubsequence(self, text1, text2):
        """
        :type text1: str
        :type text2: str
        :rtype: int
        """

        # This solution achieves the O(N) space complexity

        # We make sure that len(text1) >= len(text2): rows >= cols
        if len(text1) < len(text2):
            text1, text2 = text2, text1

        rows, cols = len(text1), len(text2)
        dp = [0] * (cols + 1)

        for i in range(1, rows + 1):
            # This represents dp[i - 1][j - 1] (the diagonal)
            prev = 0

            for j in range(1, cols + 1):
                # Save the top value before we overwrite it
                temp = dp[j]

                if text1[i - 1] == text2[j - 1]:
                    # Match: Diagonal + 1
                    dp[j] = prev + 1

                else:
                    # No match: Max of Top (dp[j]) or Left (dp[j - 1])
                    dp[j] = max(dp[j], dp[j - 1])

                # Current top becomes the diagonal for the next j
                prev = temp

        return dp[-1]
```

```python
solution = Solution()
```
