```python
class Solution(object):
    def reverseWords(self, s):
        """
        :type s: str
        :rtype: str
        """

        # Time Complexity: O(N) where n is the length of splitted
        # Space Complexity: O(N) where n is the size of lists

        # Check if s is empty
        if not s:
            return s

        # Split words with space
        splitted = s.split(" ")
        splitted.reverse()
        words = []

        for ch in splitted:
            if ch:
                words.append(ch)

        return " ".join(words)
```

```python
solution = Solution()
```

```python
s = "the sky is blue"

print(solution.reverseWords(s))
```

```python
s = "  hello world  "

print(solution.reverseWords(s))
```
