```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def deleteNode(self, root, key):
        """
        :type root: Optional[TreeNode]
        :type key: int
        :rtype: Optional[TreeNode]
        """

        # Check if there are nodes
        if not root:
            return root

        dummy = TreeNode()
        dummy.next = root

        prev = None
        del_node = None

        while root:
            if key == root.val:
                del_node = root
                break

            elif key < root.val:
                prev = root
                root = root.left

            else:
                prev = root
                root = root.right

        # If the node wasn't found, return without deletion.
        if not del_node:
            return dummy.next

        # Check if we need to delete the root
        if not prev:
            if del_node.right:
                # The right node of the root is gonna be the next root
                dummy.next = del_node.right

                head = del_node.right
                # We go though the node until the left end
                while head.left:
                    head = head.left

                head.left = del_node.left

            # Right node doesn't exist
            else:
                # Since the right node doesn't exist, we simply use
                # the left node of the root as the next root
                dummy.next = del_node.left

        # The deleted node is not the root node
        else:
            # We find if del_node is left or right to the prev node
            is_left = False

            if prev.left == del_node:
                is_left = True

            # We use the right node for the replacement for the deleted node
            # if it exists
            if del_node.right:
                # We connect the previous node to the right node
                if is_left:
                    prev.left = del_node.right

                else:
                    prev.right = del_node.right

                # We find the deepest leftmost node from the replacement node.
                head = del_node.right
                while head.left:
                    head = head.left

                # Connect the leftmost node to del_node.left
                head.left = del_node.left

            # The right node doesn't exist, so we simply connect the left node
            # to prev
            else:
                if is_left:
                    prev.left = del_node.left

                else:
                    prev.right = del_node.left

        # Disconnect the deleted node
        del_node.left = None
        del_node.right = None

        return dummy.next
```
