```python
class Solution(object):
    def dailyTemperatures(self, temperatures):
        """
        :type temperatures: List[int]
        :rtype: List[int]
        """

        answer = [0] * len(temperatures)

        # This will store indices of temperatures
        stack = []

        for current_idx, current_temp in enumerate(temperatures):
            # While the current temp is warmer than the temp at the top of the stack
            while stack and current_temp > temperatures[stack[-1]]:
                prev_idx = stack.pop()
                answer[prev_idx] = current_idx - prev_idx

            # Always push the current index onto the stack
            stack.append(current_idx)

        return answer
```

```python
solution = Solution()
```
