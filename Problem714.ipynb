```python
class Solution(object):
    def maxProfit(self, prices, fee):
        """
        :type prices: List[int]
        :type fee: int
        :rtype: int
        """

        # Maximum profit if we currently own a stock
        hold = -prices[0]

        # Maximum profit if we currently do not own a stock
        cash = 0

        for i in range(len(prices)):
            prev_hold = hold
            prev_cash = cash

            hold = max(prev_hold, prev_cash - prices[i])
            cash = max(prev_cash, prev_hold + prices[i] - fee)

        return cash
```

```python
solution = Solution()
```

```python
prices = [1,3,2,8,4,9]
fee = 2

print(solution.maxProfit(prices, fee))
```
