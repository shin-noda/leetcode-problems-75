```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def maxDepth(self, root):
        """
        :type root: Optional[TreeNode]
        :rtype: int
        """

        # Check if there are nodes
        if not root:
            return 0


        # Helper function
        def dive(depth, root):
            # Base case
            if not root:
                return depth

            return max(dive(depth + 1, root.left), dive(depth + 1, root.right))


        return max(dive(1, root.left), dive(1, root.right))
```
