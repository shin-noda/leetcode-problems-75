```python
class Solution(object):
    def letterCombinations(self, digits):
        """
        :type digits: str
        :rtype: List[str]
        """

        keypad = {
            "2": ["a", "b", "c"],
            "3": ["d", "e", "f"],
            "4": ["g", "h", "i"],
            "5": ["j", "k", "l"],
            "6": ["m", "n", "o"],
            "7": ["p", "q", "r", "s"],
            "8": ["t", "u", "v"],
            "9": ["w", "x", "y", "z"],
        }

        self.tmp = []
        self.combinations = []

        def backtrack(i):
            # Base case: Success
            if len(self.tmp) >= len(digits):
                self.combinations.append("".join(self.tmp))
                return

            chars = keypad[digits[i]]
            for ch in chars:
                # Choose
                self.tmp.append(ch)

                # Explore
                backtrack(i + 1)

                # Backtrack
                self.tmp.pop()

        backtrack(0)

        return list(self.combinations)
```

```python
solution = Solution()
```

```python
digits = "23"

print(solution.letterCombinations(digits))
```

```python
digits = "2"

print(solution.letterCombinations(digits))
```
