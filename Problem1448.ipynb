```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def goodNodes(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """

        # Check if root exists
        if not root:
            return 0

        self.count = 1

        def dive(node, max_val):
            # Base case
            if not node:
                return

            if node.val >= max_val:
                self.count += 1
                max_val = node.val

            dive(node.left, max_val)
            dive(node.right, max_val)


        dive(root.left, root.val)
        dive(root.right, root.val)

        return self.count
```
