```python
class Solution(object):
    def decodeString(self, s):
        """
        :type s: str
        :rtype: str
        """

        stack = []

        i = 0
        while i < len(s):
            if s[i] == "]":
                tmp = []
                while stack[-1].isalpha():
                    tmp.insert(0, stack.pop())

                st = "".join(tmp)

                # Pop "["
                stack.pop()

                # Get the number
                num = 0
                tmp_num = []

                # First, we have to reverse the order or the number
                while stack and stack[-1].isdigit():
                    tmp_num.append(stack.pop())

                # Then, calculate the number
                while tmp_num:
                    num = num * 10 + int(tmp_num.pop())

                # Append the string
                stack.append(st * num)

            else:
                stack.append(s[i])

            i += 1

        return "".join(stack)
```

```python
solution = Solution()
```

```python
s = "3[a]2[bc]"

print(solution.decodeString(s))
```
