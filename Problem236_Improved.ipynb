```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, x):
        self.val = x
        self.left = None
        self.right = None
```

```python
class Solution(object):
    def lowestCommonAncestor(self, root, p, q):
        """
        :type root: TreeNode
        :type p: TreeNode
        :type q: TreeNode
        :rtype: TreeNode
        """

        # We use a persistent variable to lock in the answer
        # as soon as we find it
        self.lca = None

        def dfs(node):
            # Base Case
            if not node or self.lca:
                return 0

            # Ask the left and right subtrees how many targets they found
            # There will return 0, 1, or 2
            left_finds = dfs(node.left)
            right_finds = dfs(node.right)

            # Check if the current node itslef is one of the targets (p.val or q.val)
            self_finds = 0
            if (node.val == p.val or node.val == q.val):
                self_finds = 1

            # Sum them up for this specific subtree
            total_found = left_finds + right_finds + self_finds

            # The first node (from the bottom up) that sees both targets
            # is the Lowest Common Ancestor.
            if not self.lca and total_found == 2:
                self.lca = node

            # Send the report back up to the parent
            return total_found

        # Start the cunsus from the root
        dfs(root)

        return self.lca
```
