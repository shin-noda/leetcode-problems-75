```python
class Solution(object):
    def longestSubarray(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        left = 0
        count = 0
        max_val = 0

        for right in range(len(nums)):
            if nums[right] == 0:
                count += 1

            # We only delete one element
            while count > 1:
                if nums[left] == 0:
                    count -= 1

                left += 1

            # Since we must delete one element, we don't have
            # the fence pole problem.
            max_val= max(max_val, right - left)

        return max_val
```

```python
solution = Solution()
```

```python
nums = [1,1,0,1]

print(solution.longestSubarray(nums))
```
