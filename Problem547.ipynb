```python
class Solution(object):
    def findCircleNum(self, isConnected):
        """
        :type isConnected: List[List[int]]
        :rtype: int
        """

        provinces = 0
        n = len(isConnected)
        cities = set()

        for city in range(n):
            queue = []

            # Check if this city was checked before
            if city not in cities:
                cities.add(city)
                provinces += 1
                queue.append(city)

            # We've alredy checked it
            else:
                continue

            while(queue):
                i = queue.pop()

                for j in range(n):
                    # The jth city is connected to the ith city
                    if isConnected[i][j] == 1 and j not in cities:
                        cities.add(j)

                        # We check this city later
                        queue.append(j)

        return provinces
```

```python
solution = Solution()
```

```python
isConnected = [[1,1,0],[1,1,0],[0,0,1]]

print(solution.findCircleNum(isConnected))
```
