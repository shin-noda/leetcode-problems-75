```python
class Solution(object):
    def longestOnes(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """

        # Time Complexity: O(N) amortized O(N)
        # Space Complexity: O(1)

        left = 0
        max_val = 0

        # Count how many zeros we are currently using
        count = 0

        # Use Sliding Window.
        for right in range(len(nums)):
            # We found a zero
            if nums[right] == 0:
                count += 1

            # If we've used more than k zeros, shrink from the left
            while count > k:
                if nums[left] == 0:
                    count -= 1

                left += 1

            # The window [left, right] is always valid here
            max_val = max(max_val, right - left + 1)

        return max_val
```

```python
solution = Solution()
```
