```python
class Solution(object):
    def removeStars(self, s):
        """
        :type s: str
        :rtype: str
        """

        stack = []
        for ch in s:
            if ch == "*":
                stack.pop()

            else:
                stack.append(ch)

        return "".join(stack)
```

```python
solution = Solution()
```

```python
s = "leet**cod*e"

print(solution.removeStars(s))
```

```python
s = "erase*****"

print(solution.removeStars(s))
```
