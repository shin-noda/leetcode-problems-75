```python
class Solution(object):
    def reverseVowels(self, s):
        """
        :type s: str
        :rtype: str
        """
        # Time Complexity: O(N) where N is len(s)
        # Space Complexity: O(K) where K is the size of the set.

        # Using a set for O(1) lookups
        vowels = set("aeiouAEIOU")
        s_list = list(s)
        left, right = 0, len(s) - 1

        while left < right:
            # Move left pointer until it hits a vowel
            while left < right and s_list[left] not in vowels:
                left += 1

            # Move right pointer until it hits a vowel
            while left < right and s_list[right] not in vowels:
                right -= 1

            # Swap
            s_list[left], s_list[right] = s_list[right], s_list[left]

            # Move pointers inward to avoid infinite loop
            left += 1
            right -= 1

        return "".join(s_list)
```

```python
solution = Solution()
```
