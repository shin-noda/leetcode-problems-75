```python
class Solution(object):
    def tribonacci(self, n):
        """
        :type n: int
        :rtype: int
        """

        if n == 0:
            return 0

        if n <= 2:
            return 1

        t_0 = 0
        t_1 = 1
        t_2 = 1

        t = 0
        for i in range(3, n + 1):
            t_3 = t_0 + t_1 + t_2

            # We update t's in this order to avoid overwriting values before update.
            t_0 = t_1
            t_1 = t_2
            t_2 = t_3

        return t_3
```

```python
solution = Solution()
```

```python
n = 4

print(solution.tribonacci(n))
```

```python
n = 25

print(solution.tribonacci(n))
```
