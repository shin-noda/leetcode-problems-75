```python
class Solution(object):
    def asteroidCollision(self, asteroids):
        """
        :type asteroids: List[int]
        :rtype: List[int]
        """

        stack = []

        for ast in asteroids:
            # An asteroid is moving to the right. No collision.
            if ast > 0:
                stack.append(ast)

            # An asteroid is moving to the left. May collision.
            else:
                if not stack:
                    stack.append(ast)

                else:
                    while stack:
                        prev_ast = stack.pop()

                        # No collision
                        if prev_ast < 0:
                            # Put the previous astroid and current one to the stack
                            stack.append(prev_ast)
                            stack.append(ast)
                            break

                        # Collision
                        else:
                            # Explosion: both asteroids are gone
                            if abs(ast) == prev_ast:
                                break

                            # The previous asteroid survive
                            elif abs(ast) < prev_ast:
                                stack.append(prev_ast)
                                break

                            # The current asteroid survive
                            elif not stack:
                                stack.append(ast)
                                break

        return stack
```

```python
solution = Solution()
```

```python
asteroids = [5,10,-5]

print(solution.asteroidCollision(asteroids))
```
