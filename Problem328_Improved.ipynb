```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def oddEvenList(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        # Check if head is empty.
        if not head:
            return None

        odd = head
        even = head.next

        # Save the start of the even list to join later
        even_head = even

        # We only need to check even and even.next because 'even' is always ahead.
        while even and even.next:
            # Connect current odd to the next odd (skipping the odd node)
            odd.next = even.next
            odd = odd.next

            # Connect current even to the next even (skipping the odd node)
            even.next = odd.next
            even = even.next

        # Connect the end of the odd list to the start of the even list
        odd.next = even_head

        return head
```
