```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def deleteNode(self, root, key):
        """
        :type root: Optional[TreeNode]
        :type key: int
        :rtype: Optional[TreeNode]
        """

        # Base case
        if not root:
            return None

        # Search Phase
        if key < root.val:
            root.left = self.deleteNode(root.left, key)

        elif key > root.val:
            root.right = self.deleteNode(root.right, key)

        # Deletion Phase (we found the node)
        else:
            # Case 1: No right node
            if not root.right:
                return root.left

            # Case 2: Has a right child
            # Find the leftmost leaf of the right subtree
            leftmost = root.right
            while leftmost.left:
                leftmost = leftmost.left

            # Graft the left subtree here
            leftmost.left = root.left

            # Return the right child to the parent above
            return root.right

        return root
```
