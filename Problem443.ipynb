```python
class Solution(object):
    def compress(self, chars):
        """
        :type chars: List[str]
        :rtype: int
        """
        # Note: we have to solve this problem using only constant extra space,
        # so we don't actually store results on string since it's O(N)

        result = 0

        i = 0
        while i < len(chars):
            ch = chars[i]

            if i + 1 >= len(chars) or i + 1 < len(chars) and ch != chars[i + 1]:
                # Modify the input
                chars[result] = ch
                result += 1

            else:
                streak = 1
                while i + 1 < len(chars) and ch == chars[i + 1]:
                    i += 1
                    streak += 1

                # Find the number of digits of the steak.
                # e.g. streak = 5 -> 1 digit, so add 1 to result.
                # streak = 12 -> 2 digits, so add 2 to result.
                # streak 710 -> 3 digits, so add 3 to result.

                # Update chars
                chars[result] = ch
                result += 1

                num_swap = 0
                while streak > 0:
                    chars[result] = str(streak % 10)
                    streak = streak // 10

                    # We swap number later
                    num_swap += 1

                    # We swap the numbers
                    result += 1

                # Swapping
                for j in range(num_swap // 2):
                    chars[result - num_swap + j], chars[result - 1 - j] = chars[result - 1 - j], chars[result - num_swap + j]

            i += 1

        print(chars)

        return result
```

```python
solution = Solution()
```

```python
chars = ["a","a","b","b","c","c","c"]

print(solution.compress(chars))
```
