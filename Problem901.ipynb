```python
class StockSpanner(object):

    def __init__(self):
        self.stack = []


    def next(self, price):
        """
        :type price: int
        :rtype: int
        """

        streak = 1

        print(price)

        while self.stack and self.stack[-1][0] <= price:
            item = self.stack.pop()
            streak += item[1]

        self.stack.append((price, streak))

        return streak
```
