```python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        result = nums[0]

        for i in range(1, len(nums)):
            result ^= nums[i]

        return result
```

```python
solution = Solution()
```

```python
nums = [2,2,1]

print(solution.singleNumber(nums))
```

```python
nums = [4,1,2,1,2]

print(solution.singleNumber(nums))
```
