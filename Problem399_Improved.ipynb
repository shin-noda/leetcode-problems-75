```python
class Solution(object):
    def calcEquation(self, equations, values, queries):
        """
        :type equations: List[List[str]]
        :type values: List[float]
        :type queries: List[List[str]]
        :rtype: List[float]
        """

        # Stores parent of each node
        root = {}

        # Stores weight: node / root[node] = dist[node]
        dist = {}

        def find(i):
            if i not in root:
                root[i] = i
                dist[i] = 1.0
                return i

            if root[i] == i:
                return i

            # Path Compression: recursively find the root
            last_root = find(root[i])

            # Update weight to be relative to the absolute root
            dist[i] = dist[i] * dist[root[i]]
            root[i] = last_root

            return last_root

        def union(i, j, value):
            root_i, root_j = find(i), find(j)
            if root_i != root_j:
                root[root_i] = root_j

                # i / root_i = dist[i], j / root_j = dist[j], i / j = value
                # Therefore: root_i / root_j = (i / j * j / root_j) / (i / root_i)
                dist[root_i] = (value * dist[j]) / dist[i]

        # Build the relationships
        for (i, j), val in zip(equations, values):
            union(i, j, val)

        # Process queries
        results = []
        for i, j in queries:
            if i not in root or j not in root or find(i) != find(j):
                results.append(-1.0)

            else:
                # If they share a root, i / j = (i / root) / (j / root)
                results.append(dist[i] / dist[j])

        return results
```

```python
solution = Solution()
```

```python
equations = [["a","b"],["b","c"]]
values = [2.0,3.0]
queries = [["a","c"],["b","a"],["a","e"],["a","a"],["x","x"]]

print(solution.calcEquation(equations, values, queries))
```
