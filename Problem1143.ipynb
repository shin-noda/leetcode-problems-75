```python
class Solution(object):
    def longestCommonSubsequence(self, text1, text2):
        """
        :type text1: str
        :type text2: str
        :rtype: int
        """

        # Number of rows
        rows = len(text1)

        # Number of columns
        cols = len(text2)

        # We add +1 for cols and rows for convenience
        dp = [[0] * (cols + 1) for _ in range(rows + 1)]

        for i in range(1, rows + 1):
            for j in range(1, cols + 1):
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

                if text1[i - 1] == text2[j - 1]:
                    # We add 1 to the top right cell value to account for the sequence.
                    dp[i][j] = dp[i - 1][j - 1] + 1

        return dp[-1][-1]
```

```python
solution = Solution()
```
