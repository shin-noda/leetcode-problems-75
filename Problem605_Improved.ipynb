```python
class Solution(object):
    def canPlaceFlowers(self, flowerbed, n):
        """
        :type flowerbed: List[int]
        :type n: int
        :rtype: bool
        """

        # Time Complexity: O(n)
        # Space Complexity: O(1)

        # Check if n is zero.
        if n == 0:
            return True

        i = 0
        while i <len(flowerbed):
            if flowerbed[i] == 1:
                # We skip the next plot since you cannot plan the flower next to it.
                i += 2
                continue

            if i - 1 >= 0 and flowerbed[i - 1] == 1:
                i += 1
                continue

            if i + 1 < len(flowerbed) and flowerbed[i + 1] == 1:
                i += 1
                continue

            # We modify the flowerbed
            flowerbed[i] = 1
            n -= 1

            if n == 0:
                return True

            # We skip the next plot because we planted the flower,
            # so we cannot plan it next to it.
            i += 2

        return False
```

```python
solution = Solution()
```
