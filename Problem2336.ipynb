```python
class MinHeap:
    def __init__(self):
        self.heap = []


    def insert(self, value):
        # Add to the end
        self.heap.append(value)
        self._heapify_up(len(self.heap) - 1)


    def extract_min(self):
        if not self.heap:
            return None

        if len(self.heap) == 1:
            return self.heap.pop()

        # Swap root with last
        min_val = self.heap[0]
        self.heap[0] = self.heap.pop()

        # Restore heap property
        self._heapify_down(0)

        return min_val


    def _heapify_up(self, index):
        while index > 0:
            parent = (index - 1) // 2

            if self.heap[index] < self.heap[parent]:
                self.heap[index], self.heap[parent] = self.heap[parent], self.heap[index]
                index = parent

            else:
                break


    def _heapify_down(self, index):
        size = len(self.heap)

        while True:
            left = 2 * index + 1
            right = 2 * index + 2
            smallest = index

            if left < size and self.heap[left] < self.heap[smallest]:
                smallest = left

            if right < size and self.heap[right] < self.heap[smallest]:
                smallest = right

            if smallest == index:
                break

            self.heap[index], self.heap[smallest] = self.heap[smallest], self.heap[index]
            index = smallest


    def peek(self):
        if self.heap:
            return self.heap[0]

        return None


class SmallestInfiniteSet(object):

    def __init__(self):
        self.heap = MinHeap()
        self.heap.insert(1)
        self.heap.insert(2)
        self.removed = set()


    def popSmallest(self):
        """
        :rtype: int
        """

        val = self.heap.extract_min()
        self.removed.add(val)

        if len(self.heap.heap) <= 1:
            new_val = self.heap.peek()

            while True:
                new_val += 1
                if new_val not in self.removed:
                    self.heap.insert(new_val)
                    break

        return val


    def addBack(self, num):
        """
        :type num: int
        :rtype: None
        """

        if num in self.removed:
            self.removed.remove(num)
            self.heap.insert(num)
```
