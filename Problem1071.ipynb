```python
class Solution(object):
    def gcdOfStrings(self, str1, str2):
        """
        :type str1: str
        :type str2: str
        :rtype: str
        """

        # Time Complexity: O(n^2)
        # Space Complexity: O(n)

        # Check if str1 or str2 is empty.
        if not str1 or not str2:
            return ""

        # Use hashmap for quick check
        m1 = {}
        m2 = {}

        for ch1 in str1:
            m1[ch1] = m1.get(ch1, 0) + 1

        for ch2 in str2:
            m2[ch2] = m2.get(ch2, 0) + 1

        # Check if both str1 and str2 has the same chars and order
        if m1.keys() != m2.keys():
            return ""

        # Use the lengh whicheter is shorter
        n = min(len(str1), len(str2))

        # We want the largest string, so we iterate backwards.
        for i in range(n - 1, -1, -1):
            test = str2[:i + 1]
            is_matched = True

            # Check if both str1 and str2 are multiple of test
            if len(str1) % len(test) != 0 or len(str2) % len(test) != 0:
                continue

            for j in range(0, len(str1) - len(test) + 1, len(test)):
                if str1[j:j + len(test)] != test:
                    is_matched = False
                    break

            for j in range(0, len(str2) - len(test) + 1, len(test)):
                if str2[j:j + len(test)] != test:
                    is_matched = False
                    break

            if is_matched:
                return test

        # No match
        return ""
```
