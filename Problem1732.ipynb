```python
class Solution(object):
    def largestAltitude(self, gain):
        """
        :type gain: List[int]
        :rtype: int
        """

        altitude = 0
        max_alt = 0
        for val in gain:
            altitude += val
            max_alt = max(max_alt, altitude)

        return max_alt
```

```python
solution = Solution()
```

```python
gain = [-5,1,5,0,-7]

print(solution.largestAltitude(gain))
```
