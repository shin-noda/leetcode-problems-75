```python
class Solution(object):
    def asteroidCollision(self, asteroids):
        """
        :type asteroids: List[int]
        :rtype: List[int]
        """

        stack = []

        for ast in asteroids:
            # The collision loop
            while stack and ast < 0 and stack[-1] > 0:
                diff = ast + stack[-1]

                if diff < 0:
                    # Current (left) is bigger: pop stack and keep colliding
                    stack.pop()
                    continue

                elif diff == 0:
                    # Both are the same size: both explode, move to next asteroid
                    stack.pop()

                    # Mark current as gone
                    ast = 0

                else:
                    # Previous (right) is bigger: current explodes
                    ast = 0

                break

            # If current asteroid survived the loop (or never needed to collide)
            if ast:
                stack.append(ast)

        return stack
```

```python
solution = Solution()
```

```python
asteroids = [5,10,-5]

print(solution.asteroidCollision(asteroids))
```
