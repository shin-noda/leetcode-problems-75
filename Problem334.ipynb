```python
class Solution(object):
    def increasingTriplet(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """

        # Check if there is at least 3 numbers in nums
        if len(nums) < 3:
            return False

        # This is a longest increasing sequence problem,
        # so we don't keep track of real three numbers but the potential.
        # Thus, in the end, the 3 numbers does not represent the real sequence.

        i = float('inf')
        j = float('inf')
        k = float('inf')

        for num in nums:
            if num <= i:
                i = num

            else:
                if num <= j:
                    j = num

                else:
                    if num <= k:
                        k = num
                        return True

        return False
```

```python
solution = Solution()
```

```python
nums = [1,2,3,4,5]

print(solution.increasingTriplet(nums))
```
