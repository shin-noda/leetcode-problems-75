```python
class Solution(object):
    def nearestExit(self, maze, entrance):
        """
        :type maze: List[List[str]]
        :type entrance: List[int]
        :rtype: int
        """
        # Use DFS

        steps = 0
        queue = [(entrance[0], entrance[1])]
        path = set()

        while queue:
            n = len(queue)
            tmp = []

            for _ in range(n):
                r, c = queue.pop()

                if (
                    (r, c) in path or
                    r < 0 or r > len(maze) - 1 or
                    c < 0 or c > len(maze[0]) - 1 or
                    maze[r][c] == "+"
                ):
                    continue

                # Found the exit
                if (
                    steps != 0 and (
                    r == 0 or r == len(maze) - 1 or
                    c == 0 or c == len(maze[0]) - 1)
                ):
                    return steps

                path.add((r, c))

                # Add Up, Down, Left, and Right
                tmp.append((r - 1, c))
                tmp.append((r + 1, c))
                tmp.append((r, c - 1))
                tmp.append((r, c + 1))

            queue = tmp
            steps += 1

        return -1
```

```python
solution = Solution()
```

```python
maze = [["+","+",".","+"],[".",".",".","+"],["+","+","+","."]]
entrance = [1,2]

print(solution.nearestExit(maze, entrance))
```
