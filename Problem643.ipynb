```python
class Solution(object):
    def findMaxAverage(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: float
        """

        # Time Complexity: O(N) where n is the lenght of nums
        # Space Complexity: O(1)

        # Check if k is less than the length of nums
        if k > len(nums):
            return 0.0

        left = 0

        # We need those 2 variables to keep track to the current and the max values.
        sum_val = 0
        curr = 0

        for i in range(len(nums)):
            # Simply calculate the sum from i = 0 to i = k - 1
            if i < k:
                sum_val += nums[i]
                curr = sum_val

            # Use Sliding Window to calculate the max average subarray
            else:
                curr += nums[i] - nums[left]
                sum_val = max(sum_val, curr)
                left += 1

        # Note: float(sum_val / k) did not work as intended.
        return float(sum_val) / k
```

```python
solution = Solution()
```

```python
nums = [1,12,-5,-6,50,3]
k = 4

print(solution.findMaxAverage(nums, k))
```
