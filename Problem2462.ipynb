```python
class MinHeap:
    def __init__(self):
        self.heap = []


    def insert(self, value):
        self.heap.append(value)

        # Sift up
        self._heapify_up(len(self.heap) - 1)


    def extract_min(self):
        if not self.heap:
            return None

        if len(self.heap) == 1:
            return self.heap.pop()

        min_val = self.heap[0]
        self.heap[0] = self.heap.pop()

        # Sift down
        self._heapify_down(0)

        return min_val


    def _heapify_up(self, index):
        while index > 0:
            parent = (index - 1) // 2

            if self.heap[parent] > self.heap[index]:
                self.heap[parent], self.heap[index] = self.heap[index], self.heap[parent]
                index = parent

            else:
                break


    def _heapify_down(self, index):
        size = len(self.heap)

        while True:
            left = 2 * index + 1
            right = 2 * index + 2
            smallest = index

            if left < size and self.heap[smallest] > self.heap[left]:
                smallest = left

            if right < size and self.heap[smallest] > self.heap[right]:
                smallest = right

            if smallest == index:
                break

            self.heap[smallest], self.heap[index] = self.heap[index], self.heap[smallest]
            index = smallest


    def peek(self):
        if self.heap:
            return self.heap[0]

        return None


class Solution(object):
    def totalCost(self, costs, k, candidates):
        """
        :type costs: List[int]
        :type k: int
        :type candidates: int
        :rtype: int
        """

        first_heap = MinHeap()
        last_heap = MinHeap()
        total_cost = 0
        left = 0
        right = len(costs) - 1

        # We set up the initial min_heap
        for _ in range(candidates):
            if left <= right:
                first_heap.insert(costs[left])
                left += 1

            # We have to check it twice since left may be incremented.
            if left <= right:
                last_heap.insert(costs[right])
                right -= 1

            # No more candidate left in candidates
            else:
                break

        for i in range(k):
            # Boths heaps are non-empty
            if first_heap.heap and last_heap.heap:
                if first_heap.peek() <= last_heap.peek():
                    total_cost += first_heap.extract_min()

                    # We add a new candidate from the beginning of costs
                    if left <= right:
                        first_heap.insert(costs[left])
                        left += 1

                else:
                    total_cost += last_heap.extract_min()

                    # We add a new candidate from the end of costs
                    if left <= right:
                        last_heap.insert(costs[right])
                        right -= 1

            # last_heap is empty
            elif first_heap.heap:
                total_cost += first_heap.extract_min()

            # first_heap is empty
            else:
                total_cost += last_heap.extract_min()

        return total_cost
```

```python
solution = Solution()
```
