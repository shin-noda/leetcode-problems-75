```python
class Solution(object):
    def minCostClimbingStairs(self, cost):
        """
        :type cost: List[int]
        :rtype: int
        """

        # We just track the last two best costs
        prev2 = cost[0]
        prev1 = cost[1]

        for i in range(2, len(cost)):
            current = cost[i] + min(prev1, prev2)
            prev2 = prev1
            prev1 = current

        # The top floor is reached from either of the last two steps
        return min(prev1, prev2)
```

```python
solution = Solution()
```

```python
cost = [10,15,20]

print(solution.minCostClimbingStairs(cost))
```

```python
cost = [1,100,1,1,1,100,1,1,100,1]

print(solution.minCostClimbingStairs(cost))
```
