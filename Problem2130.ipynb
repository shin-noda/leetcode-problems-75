```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def pairSum(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: int
        """

        # Check if the nodes exist
        if not head or not head.next:
            return head

        # Use the fast and slow pointer
        slow = head
        fast = head.next

        nums = [slow.val]
        max_val = float('-inf')

        # We collect the half of nums
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next

            nums.append(slow.val)

        # Iterate the remaining half to calculate the twin sum
        for i in range(len(nums)):
            slow = slow.next
            max_val = max(max_val, slow.val + nums[-(1 + i)])

        return max_val
```
