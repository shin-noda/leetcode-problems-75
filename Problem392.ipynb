```python
class Solution(object):
    def isSubsequence(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """

        # Use a two-pointer method.
        p1 = 0
        p2 = 0

        n1 = len(s)
        n2 = len(t)

        while p1 < n1 and p2 < n2:
            # Characters match
            if s[p1] == t[p2]:
                p1 += 1

            p2 += 1

        return p1 == n1
```

```python
solution = Solution()
```

```python
s = "abc"
t = "ahbgdc"

print(solution.isSubsequence(s, t))
```
