```python
class Solution(object):
    def orangesRotting(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: int
        """

        stack = []
        rows, cols = len(grid), len(grid[0])

        # Count number of fresh orange in the grid
        orange = 0
        for i in range(rows):
            for j in range(cols):
                # Fresh orange
                if grid[i][j] == 1:
                    orange += 1

                # Rotten orange
                elif grid[i][j] == 2:
                    stack.append((i, j))

        # Check if there are fresh orange in the grid
        if orange == 0:
            return 0

        minutes = 0
        while stack:
            minutes += 1
            n = len(stack)
            next_level = []

            for _ in range(n):
                r, c = stack.pop()

                for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                    nr, nc = r + dr, c + dc

                    # Check if a fresh orange is in the next grid
                    if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 1:
                        orange -= 1
                        grid[nr][nc] = 2

                        # Check if any fresh orange left
                        if orange == 0:
                            return minutes

                        next_level.append((nr, nc))

            stack = next_level

        return -1
```

```python
solution = Solution()
```

```python
grid = [[2,1,1],[1,1,0],[0,1,1]]

print(solution.orangesRotting(grid))
```
