```python
from collections import deque

class Solution(object):
    def eraseOverlapIntervals(self, intervals):
        """
        :type intervals: List[List[int]]
        :rtype: int
        """

        # Remove duplicates
        tmp = set(tuple(x) for x in intervals)

        # Convert back to the list because we care about the order
        # and we remove some elements later
        inter = [list(x) for x in tmp]

        # Sort the list
        inter.sort()

        results = []
        dq = deque(inter)

        prev_start, prev_end = dq.popleft()

        # We do this for the edge case when
        # there is one element in dq
        curr_start, curr_end = prev_start, prev_end

        while True:
            # We don't do while dq because we want to append the last element
            # of the dq
            if not dq:
                results.append([curr_start, curr_end])
                break

            curr_start, curr_end = dq.popleft()

            # If the previous interval is like (1,2) and
            # the new one is (1,3), they are overlapping
            if curr_start == prev_start:
                # Discard the current interval from the set (do nothing)
                continue

            # The starting point is different
            else:
                # Check if it's overlapping
                # Note: it's okay to touch prev_end and curr_start,
                # so that's why it's '<' instead of '<='.
                if curr_start < prev_end:
                    # We remove the new interval
                    if prev_end < curr_end:
                        continue

                    # We remove the previous interval
                    else:
                        prev_start = curr_start
                        prev_end = curr_end

                # No overlapping
                else:
                    results.append([prev_start, prev_end])

                    # Update the previous interval
                    prev_start = curr_start
                    prev_end = curr_end

        return len(intervals) - len(results)
```

```python
solution = Solution()
```

```python
intervals = [[1,2],[2,3],[3,4],[1,3]]

print(solution.eraseOverlapIntervals(intervals))
```
