```python
class Solution(object):
    def productExceptSelf(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """

        # Use prefix and suffix to solve the problem
        results = [1] * len(nums)
        right_val = 1

        for i in range(1, len(nums)):
            results[i] = results[i - 1] * nums[i - 1]

        for i in range(len(nums) - 2, -1, -1):
            # We need right_val so that
            # we don't mistakenly alter the previous calculations.
            right_val *= nums[i + 1]
            results[i] *= right_val

        return results
```

```python
solution = Solution()
```

```python
nums = [1,2,3,4]

print(solution.productExceptSelf(nums))
```

```python
nums = [-1,1,0,-3,3]

print(solution.productExceptSelf(nums))
```

```python
nums = [2,1]

print(solution.productExceptSelf(nums))
```
