```python
# Definition for singly-linked list.
class ListNode(object):
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
class Solution(object):
    def deleteMiddle(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """

        # A sentinel node
        dummy = ListNode()
        dummy.next = head

        n = 0
        while head:
            head = head.next
            n += 1

        half = n // 2

        # Rest the head
        head = dummy
        for _ in range(half):
            head = head.next

        # Check if there is a two nodes ahead of the current node
        if head.next and head.next.next:
            head.next = head.next.next

        else:
            head.next = None

        return dummy.next
```
