```python
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
class Solution(object):
    def pathSum(self, root, targetSum):
        """
        :type root: Optional[TreeNode]
        :type targetSum: int
        :rtype: int
        """

        # Check if there are nodes
        if not root:
            return 0

        self.count = 0

        def dive(node, sums):
            # Base case
            if not node:
                return

            if node.val == targetSum:
                self.count += 1

            for i in range(len(sums)):
                if sums[i] + node.val == targetSum:
                    self.count += 1

                sums[i] += node.val

            sums.append(node.val)

            dive(node.left, list(sums))
            dive(node.right, list(sums))

        dive(root, [])

        return self.count
```
