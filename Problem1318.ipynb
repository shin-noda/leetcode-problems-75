```python
class Solution(object):
    def minFlips(self, a, b, c):
        """
        :type a: int
        :type b: int
        :type c: int
        :rtype: int
        """

        # We want to flip bits of a and b so that
        # a | b == c

        # Current reuslt of a | b
        ab = a | b

        # We ignore the first "0b" part
        b_a = bin(a)[2:]
        b_b = bin(b)[2:]
        b_ab = bin(ab)[2:]
        b_c = bin(c)[2:]

        if len(b_ab) > len(b_c):
            b_c = "0" * (len(b_ab) - len(b_c)) + b_c

        elif len(b_ab) < len(b_c):
            b_ab = "0" * (len(b_c) - len(b_ab)) + b_ab

        # Add zero's to b_a and b_b
        if len(b_a) < len(b_c):
            b_a = "0" * (len(b_c) - len(b_a)) + b_a

        if len(b_b) < len(b_c):
            b_b = "0" * (len(b_c) - len(b_b)) + b_b

        count = 0
        n = len(b_ab)
        for i in range(n):
            if b_ab[i] != b_c[i]:
                if b_c[i] == "1":
                    # We only need to flip one bit from either a or b
                    count += 1

                else:
                    # We many need to flip both a and b
                    if b_a[i] == "1":
                        count += 1

                    if b_b[i] == "1":
                        count += 1

        return count
```

```python
solution = Solution()
```
