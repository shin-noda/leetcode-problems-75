```python
class Solution(object):
    def increasingTriplet(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """

        # Check if there is at least 3 numbers in nums
        if len(nums) < 3:
            return False

        # These track the absolute smallest values seen
        first = float('inf')
        second = float('inf')
        third = float('inf')

        # This locks in the partner for our middle value
        best_first_for_second = None

        for num in nums:
            if num <= first:
                # We found a new "smallest"
                first = num

            elif num <= second:
                # We found a middle number better than our current one.
                second = num
                best_first_for_second = first

            else:
                # If we are here, num > second.
                # Since we locked in best_first_for_second,
                # We have a guarantee valid i < j < k sequence
                third = num

                # [best_first_for_second, second, num]

                return True

        return False
```

```python
solution = Solution()
```
