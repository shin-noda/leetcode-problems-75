```python
class Solution(object):
    def nearestExit(self, maze, entrance):
        """
        :type maze: List[List[str]]
        :type entrance: List[int]
        :rtype: int
        """

        # Use Breadth-First Search

        rows, cols = len(maze), len(maze[0])

        # Start queue with (row, col, current_distance)
        queue = [(entrance[0], entrance[1], 0)]

        # Mark entrance as visited by changing it to a wall
        # This saves memory by not using a separate set
        maze[entrance[0]][entrance[1]] = "+"

        # Process the queue level by level
        while queue:
            next_level = []
            for r, c, dist in queue:
                # Explore 4 directions: Up, Down, Left, Right
                for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                    nr, nc = r + dr, c + dc

                    # Check boundaries and if it's a path
                    if 0 <= nr < rows and 0 <= nc < cols and maze[nr][nc] == ".":
                        # Check if this new cell is an exit
                        if nr == 0 or nr == rows - 1 or nc == 0 or nc == cols - 1:
                            return dist + 1

                        # Mark as visited immediately and add to next level
                        maze[nr][nc] = "+"
                        next_level.append((nr, nc, dist + 1))

            queue = next_level

        return -1
```

```python
solution = Solution()
```

```python
maze = [["+","+",".","+"],[".",".",".","+"],["+","+","+","."]]
entrance = [1,2]

print(solution.nearestExit(maze, entrance))
```
