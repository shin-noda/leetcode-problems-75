```python
class Solution(object):
    def countBits(self, n):
        """
        :type n: int
        :rtype: List[int]
        """

        # Check if n is 0
        if n == 0:
            return [0]

        map = {0: 0, 1: 1}
        denominator = 2

        for i in range(2, n + 1):
            if i // denominator == 0:
                map[i] = 1

                # Multiply by 2
                denominator *= 2

            else:
                map[i] = map[i // denominator] + map[i % denominator]

        return list(map.values())
```

```python
solution = Solution()
```

```python
n = 2

print(solution.countBits(n))
```
