```python
class Solution(object):
    def maxOperations(self, nums, k):
        """
        :type nums: List[int]
        :type k: int
        :rtype: int
        """

        # Time Complexity: O(N log N) for Tim Sort

        operations = 0
        n = len(nums)
        m = {}

        # Sort nums in a reverse order
        nums.sort(reverse=True)

        # There are no negative numbers, so we skip nums
        # until nums[i] < k.

        i = 0
        while i < len(nums) and nums[i] >= k:
            i += 1

        # Use a hashmap to count frequencies of numbers
        for i in range(len(nums)):
            num = nums[i]
            m[num] = m.get(num, 0) + 1

        for key in m:
            # We calculate the compliment key
            v1 = m.get(key)

            if k - key in m:
                v2 = m.get(k - key)

                # Check if key and the compliment values are the same
                if key == k - key:
                    operations += v1 // 2
                    m[key] = 0

                else:
                    operations += min(v1, v2)
                    m[key] = 0
                    m[k - key] = 0

            # Compliment key doesn't exist
            else:
                m[key] = 0

        return operations
```
