```python
class Solution(object):
    def uniquePaths(self, m, n):
        """
        :type m: int
        :type n: int
        :rtype: int
        """

        # Create a m x n matrix
        dp = [[0] * n for _ in range(m)]

        # Fill the first row
        for i in range(n):
            dp[0][i] = 1

        # Fill the first column
        for i in range(m):
            dp[i][0] = 1

        # Fill the lest of the cells
        for i in range(1, m):
            for j in range(1, n):
                # We can go there from the top or from the left
                dp[i][j] = dp[i - 1][j] + dp[i][j - 1]

        return dp[-1][-1]
```

```python
solution = Solution()
```

```python
m = 3
n = 7

print(solution.uniquePaths(m, n))
```
