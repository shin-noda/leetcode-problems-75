```python
class Solution(object):
    def canPlaceFlowers(self, flowerbed, n):
        """
        :type flowerbed: List[int]
        :type n: int
        :rtype: bool
        """

        # Time Complexity: O(n)
        # Space Complexity: O(1)

        # Check if n is zero.
        if n == 0:
            return True

        for i in range(len(flowerbed)):
            if flowerbed[i] == 1:
                continue

            if i - 1 >= 0 and flowerbed[i - 1] == 1:
                continue

            if i + 1 < len(flowerbed) and flowerbed[i + 1] == 1:
                continue

            # We modify the flowerbed
            flowerbed[i] = 1
            n -= 1

            if n == 0:
                return True

        return False
```

```python
solution = Solution()
```

```python
flowerbed = [1,0,0,0,1]
n = 1

print(solution.canPlaceFlowers(flowerbed, n))
```
