```python
class Solution(object):
    def minEatingSpeed(self, piles, h):
        """
        :type piles: List[int]
        :type h: int
        :rtype: int
        """

        # Define the search space for k (the speed)
        # The slowest she can eat is 1 banana/hr
        # The fastest she needs to eas is the largest pile/hr
        # -> 1 <= k <= max(piles)

        left = 1
        right = max(piles)
        answer = right

        # Binary search through the values of k
        while left <= right:
            mid = (left + right) // 2

            # Calculate total hours for speed mid
            total_hours = 0
            for pile in piles:
                total_hours += pile // mid
                if pile % mid != 0:
                    total_hours += 1

            # Adjust the search range
            if total_hours <= h:
                # This speed works
                # Record it and try to find an even slower speed (smaller k)
                answer = mid
                right = mid - 1

            else:
                # Too slow. We need a higher speed
                left = mid + 1

        return answer
```

```python
solution = Solution()
```
