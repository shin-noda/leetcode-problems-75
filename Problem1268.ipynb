```python
class TrieNode(object):
    def __init__(self):
        self.nodes = [None] * 26
        self.is_end = False


class Trie(object):
    def __init__(self):
        self.dummy = TrieNode()
        self.count = 0


    def _char_to_int(self, ch):
        return ord(ch) - ord('a')


    def _int_to_char(self, num):
        return chr(ord('a') + num)


    def _dfs(self, chars, head, words):
        # Base case
        if not head:
            return

        # Success
        if self.count >= 3:
            return

        if head.is_end:
            words.append("".join(chars))
            self.count += 1

        nodes = head.nodes

        for i in range(len(nodes)):
            node = nodes[i]

            if node:
                # Choose
                ch = self._int_to_char(i)
                chars.append(ch)

                # Explore
                self._dfs(chars, node, words)

                # Unchoose
                chars.pop()


    def insert(self, word):
        head = self.dummy

        for w in word:
            idx = self._char_to_int(w)

            if not head.nodes[idx]:
                head.nodes[idx] = TrieNode()

            head = head.nodes[idx]

        # This is the end node
        head.is_end = True


    def startsWith(self, prefix):
        head = self.dummy

        chars = []
        results = []
        for p in prefix:
            chars.append(p)

            idx = self._char_to_int(p)

            if head:
                head = head.nodes[idx]

            tmp = []
            self.count = 0
            self._dfs(list(chars), head, tmp)
            results.append(tmp)

        return results


class Solution(object):
    def suggestedProducts(self, products, searchWord):
        """
        :type products: List[str]
        :type searchWord: str
        :rtype: List[List[str]]
        """
        trie = Trie()

        for product in products:
            trie.insert(product)

        return trie.startsWith(searchWord)
```

```python
solution = Solution()
```

```python
products = ["mobile","mouse","moneypot","monitor","mousepad"]
searchWord = "mouse"

print(solution.suggestedProducts(products, searchWord))
```
