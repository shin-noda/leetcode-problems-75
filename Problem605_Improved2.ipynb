```python
class Solution(object):
    def canPlaceFlowers(self, flowerbed, n):
        """
        :type flowerbed: List[int]
        :type n: int
        :rtype: bool
        """

        # Time Complexity: O(n)
        # Space Complexity: O(1)

        if n == 0:
            return True

        i = 0
        m = len(flowerbed)

        while i < m:
            if flowerbed[i] == 0:
                # Check if we are at the end or the next one is empty
                if i == m - 1 or flowerbed[i + 1] == 0:
                    n -= 1

                    if n == 0:
                        return True

                    # We planted, so skip the next one
                    i += 2

                else:
                    # Current one is 0 but next one is 1,
                    # so skip the current, the next 1, and the 1's neighbour.
                    i += 3

            else:
                # Curret one is 1, so skip it's neighbour.
                i += 2

        return n <= 0
```

```python
solution = Solution()
```
