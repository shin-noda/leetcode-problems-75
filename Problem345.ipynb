```python
class Solution(object):
    def reverseVowels(self, s):
        """
        :type s: str
        :rtype: str
        """
        # Time Complexity: O(N) where N is len(s)
        # Space Complexity: O(N) where N is the length of s_list

        # Check is s is empty or its length is 1.
        if not s or len(s) == 1:
            return s

        vowels = ['a', 'e', 'i', 'o', 'u']
        indices = []
        n = len(s)
        s_list = list(s)

        for i in range(n):
            if s[i].lower() in vowels:
                indices.append(i)

        for i in range(len(indices) // 2):
            # Swap i1 and i2
            i1 = indices[i]
            i2 = indices[len(indices) - 1 - i]

            s_list[i1], s_list[i2] = s_list[i2], s_list[i1]

        return "".join(s_list)
```

```python
solution = Solution()
```

```python
s = "IceCreAm"

print(solution.reverseVowels(s))
```
