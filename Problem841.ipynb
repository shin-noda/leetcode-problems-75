```python
class Solution(object):
    def canVisitAllRooms(self, rooms):
        """
        :type rooms: List[List[int]]
        :rtype: bool
        """

        # Check if there are no rooms or one room
        if len(rooms) <= 1:
            return True

        key_set = set()
        room_set = set()

        for i in range(len(rooms)):
            # Add the room number
            room_set.add(i)

        # Open the first room to get keys
        queue = [0]
        queue.extend(rooms[0])
        key_set.add(0)

        # We opened the first room
        room_set.remove(0)

        while queue:
            key = queue.pop()

            # Update the set
            if key not in key_set:
                key_set.add(key)

                # Open the room
                room_set.discard(key)

                # Update queue
                queue.extend(rooms[key])

        # If no room left in the set, we opened all rooms
        return not room_set
```

```python
solution = Solution()
```

```python
rooms = [[1],[2],[3],[]]

print(solution.canVisitAllRooms(rooms))
```

```python
rooms = [[1,3],[3,0,1],[2],[0]]

print(solution.canVisitAllRooms(rooms))
```
