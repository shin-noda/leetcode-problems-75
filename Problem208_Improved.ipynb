```python
class TrieNode(object):
    def __init__(self):
        self.nodes = [None] * 26
        self.is_end = False


class Trie(object):

    def __init__(self):
        self.dummy = TrieNode()


    def _char_to_int(self, ch):
        return int(ord(ch) - ord('a'))


    def insert(self, word):
        """
        :type word: str
        :rtype: None
        """

        head = self.dummy
        for w in word:
            if head.nodes[self._char_to_int(w)]:
                head = head.nodes[self._char_to_int(w)]

            else:
                head.nodes[self._char_to_int(w)] = TrieNode()
                head = head.nodes[self._char_to_int(w)]

        # This is the last word
        head.is_end = True


    def search(self, word):
        """
        :type word: str
        :rtype: bool
        """

        head = self.dummy
        for w in word:
            if head.nodes[self._char_to_int(w)]:
                head = head.nodes[self._char_to_int(w)]

            else:
                return False

        return head.is_end


    def startsWith(self, prefix):
        """
        :type prefix: str
        :rtype: bool
        """

        head = self.dummy
        for p in prefix:
            if head.nodes[self._char_to_int(p)]:
                head = head.nodes[self._char_to_int(p)]

            else:
                return False

        return True
```
