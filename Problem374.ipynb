```python
"""
# The guess API is already defined for you.
@param num, your guess
@return -1 if num is higher than the picked number
         1 if num is lower than the picked number
         otherwise return 0
def guess(num):
"""
```

```python
class Solution(object):
    def guessNumber(self, n):
        """
        :type n: int
        :rtype: int
        """

        # We use a slightly modifed binary search because
        # we need to use the custom method, guess(num)

        left = 1
        right = n

        while left <= right:
            mid = (left + right) // 2

            answer = guess(mid)

            if answer == 0:
                return mid

            elif answer == -1:
                right = mid - 1

            else:
                left = mid + 1
```
