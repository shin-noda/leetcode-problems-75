```python
class Solution(object):
    def rob(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # Check if there is only one house
        if len(nums) == 1:
            return nums[0]

        # Check if there are only two houses
        if len(nums) == 2:
            return max(nums[0], nums[1])

        dp = [0] * len(nums)

        # Initialize first two elements
        dp[0] = nums[0]
        dp[1] = max(nums[0], nums[1])

        for i in range(2, len(nums)):
            dp[i] = max(dp[i - 1], nums[i] + dp[i - 2])

        return dp[-1]
```

```python
solution = Solution()
```

```python
nums = [1,2,3,1]

print(solution.rob(nums))
```

```python
nums = [2,7,9,3,1]

print(solution.rob(nums))
```
